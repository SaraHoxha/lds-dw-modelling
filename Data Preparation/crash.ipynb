```python
from utils.read_write import read_csv_v2, read_csv_v3
```

```python
CRASHES_DF = read_csv_v2('../data/Crashes_Processed.csv')
```

```python
foreignKeys = ["Crash_Date_ID", "Police_Notified_Date_ID", "Crash_Location_ID", "Crash_Condition_ID", "Injury_ID"]
normalKeys = ["Primary_Contributory_Cause", "Secondary_Contributory_Cause", "Number_of_Units", "Most_Severe_Injury", "Difference_Between_Crash_Date_And_Police_Notified"]
```

```python
dateTimeTable = read_csv_v3('dw_tables_csv/DateTime.csv', 'DateTime_ID')
```

```python
crashLocationTable = read_csv_v3('dw_tables_csv/CrashLocation.csv', 'Crash_Location_ID')
```

```python
crashConditionTable = read_csv_v3('dw_tables_csv/CrashCondition.csv', 'Crash_Condition_ID')
```

```python
crashInjuriesTable = read_csv_v3('dw_tables_csv/CrashInjury.csv', 'Crash_Injury_ID')
```

```python
def findIDFromTable (row, columns, table, tablePrimaryKey):
    column_values = [str(row[col.upper()]) for col in columns]
    

    key = " ".join(column_values).strip()
    try:
        row_data = table[key]
        return row_data[tablePrimaryKey]
    except Exception as e:
        print (columns)
        print (column_values)
        print (key)
```

```python
result = []
for index, row in enumerate(CRASHES_DF):
    newRow = {
        "Crash_ID": index,
        "Primary_Contributory_Cause": row["PRIM_CONTRIBUTORY_CAUSE"],
        "Secondary_Contributory_Cause": row["SEC_CONTRIBUTORY_CAUSE"],
        "Number_of_Units": row["NUM_UNITS"],
        "Most_Severe_Injury": row["MOST_SEVERE_INJURY"],
        "Difference_Between_Crash_Date_And_Police_Notified": row["DELTA_TIME_CRASH_DATE_POLICE_REPORT_DATE"],
    }
    newRow["Crash_Date_ID"] = findIDFromTable(row, ['DAY', 'MONTH', 'YEAR', 'TIME'], dateTimeTable, "DateTime_ID")
    newRow["Police_Notified_Date_ID"] = findIDFromTable(row, ['DAY_POLICE_NOTIFIED', 'MONTH_POLICE_NOTIFIED', 'YEAR_POLICE_NOTIFIED', 'TIME_POLICE_NOTIFIED'], dateTimeTable, "DateTime_ID")
    newRow["Crash_Condition_ID"] = findIDFromTable(row, ['Posted_Speed_Limit','Traffic_Control_Device', 'Device_Condition', 'Weather_Condition', 'Lighting_Condition', 'Trafficway_Type', 'Roadway_Surface_Cond', 'Road_Defect', 'Alignment'], crashConditionTable, "Crash_Condition_ID")
    newRow["Crash_Injury_ID"] = findIDFromTable(row, ['Injuries_Total', 'Injuries_Fatal', 'Injuries_Incapacitating', 'Injuries_Non_Incapacitating', 'Injuries_No_Indication', 'Injuries_Reported_Not_Evident', 'Injuries_Unknown'], crashInjuriesTable, "Crash_Injury_ID")
    newRow["Crash_Location_ID"] = findIDFromTable(row, ['Street_Name', 'Street_No', 'Street_Direction', 'Beat_Of_Occurrence', 'Latitude', 'Longitude', 'Location'], crashLocationTable, "Crash_Location_ID")

    result.append (newRow)
```
