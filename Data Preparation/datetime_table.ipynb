```python
from utils.read_write import read_csv_v2, to_csv_v2
```

```python
processedFileString = '_Processed.csv'
people_df=read_csv_v2('../data/People'+processedFileString)
vehicles_df=read_csv_v2('../data/Vehicles'+processedFileString)
crashes_df=read_csv_v2('../data/Crashes'+processedFileString)
```

```python
from utils.utils import select_columns
```

```python
filtered_ppl = select_columns (people_df, ['YEAR', 'MONTH', 'DAY', 'TIME'])
```

```python
filtered_ppl
```

```python
filtered_vehicl = select_columns (vehicles_df, ['YEAR', 'MONTH', 'DAY', 'TIME'])
```

```python
filtered_vehicl
```

```python
filtered_crashes = select_columns (crashes_df, ['YEAR', 'MONTH', 'DAY', 'TIME'])
```

```python
filtered_crashes
```

```python
allDates = filtered_ppl + filtered_vehicl + filtered_crashes
```

```python
dateTimeTable = {}
```

```python
from utils.utils import concatenate_values
```

```python
dateTimeID = 1
for date in allDates:
    key = concatenate_values (date)
    if key not in dateTimeTable:
        dateTimeTable[key] = {"DateTime_ID": dateTimeID, "Day":date['DAY'], "Month":date["MONTH"], "Year":date["YEAR"], "Time":date["TIME"]}
        dateTimeID += 1
```

```python
from utils.read_write import to_csv
```

```python
to_csv(dateTimeTable, "dw_tables_csv/dateTime.csv")
```
